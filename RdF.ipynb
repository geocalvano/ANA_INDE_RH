```python
import csv
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, FOAF
from rdflib.namespace import XSD

# Define o namespace para os recursos e predicados personalizados
ex = Namespace("http://example.org/")

# Carrega o arquivo CSV e cria um grafo RDF
graph = Graph()

# Abra o arquivo CSV e leia a primeira linha (cabeçalho)
csv_file_path = "/content/TESTE_CSV_8.csv"
header = None
data = []
with open(csv_file_path, newline='', encoding='utf-8-sig') as csvfile:
    reader = csv.DictReader(csvfile, delimiter=';')
    header = reader.fieldnames
    data = [row for row in reader]  # Armazena as linhas do CSV em uma lista

# Verifica se a coluna "ORDEM" existe no cabeçalho do CSV
if 'ORDEM' not in header:
    print("Coluna 'ORDEM' não encontrada no arquivo CSV.")
else:
    # Processa as linhas do CSV armazenadas na lista
    for row in data:
        # Cria um recurso RDF usando o valor da coluna "ORDEM" como identificador
        subject = ex[row['ORDEM']]
        graph.add((subject, RDF.type, FOAF.Person))

        # Adiciona predicados e objetos ao recurso
        graph.add((subject, ex.ORDEM, Literal(row['ORDEM'])))
        graph.add((subject, ex.TERMO, Literal(row['TERMO'])))
        graph.add((subject, ex.ND, Literal(row['ND'])))

# Serializa o grafo RDF em Turtle e salva em um arquivo
output_file_path = "/content/saida.ttl"
graph.serialize(destination=output_file_path, format='turtle')
```
